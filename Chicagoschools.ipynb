```python
#What are the top 10 schools with the highest "Average Student Attendance"?
%sql select school_id, name_of_school,AVERAGE_STUDENT_ATTENDANCE FROM schools order by AVERAGE_STUDENT_ATTENDANCE desc nulls last limit 10 
```

```python
#Retrieve the list of 5 Schools with the lowest Average Student Attendance sorted in ascending order based on attendance
%sql select school_id, name_of_school,AVERAGE_STUDENT_ATTENDANCE FROM schools order by AVERAGE_STUDENT_ATTENDANCE asc limit 5
```

```python
#Which Schools have Average Student Attendance lower than 70%?
%sql select school_id,name_of_school,AVERAGE_STUDENT_ATTENDANCE FROM schools where CAST(REPLACE(Average_Student_Attendance, '%', '') AS DOUBLE ) < 70\
order by AVERAGE_STUDENT_ATTENDANCE 
```

```python
#Get the total College Enrollment for each Community Area
%sql select SUM(COLLEGE_ENROLLMENT) AS Total_Enrollment,COMMUNITY_AREA_NAME\
FROM SCHOOLS group by COMMUNITY_AREA_NAME
```

```python
#Get the 5 Community Areas with the least total College Enrollment sorted in ascending order
%sql select school_id,name_of_school,COLLEGE_ENROLLMENT,COMMUNITY_AREA_NAME FROM schools where CAST(REPLACE(Average_Student_Attendance, '%', '') AS DOUBLE ) < 70\
order by COLLEGE_ENROLLMENT limit 5
```

```python
#List 5 schools with lowest safety score.
%sql select NAME_OF_SCHOOL,SAFETY_SCORE FROM SCHOOLS ORDER BY SAFETY_SCORE NULLS LAST LIMIT 5
```
